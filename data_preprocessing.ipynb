daily_activity contains all the columns of the merged dataframe with a few column name differences, so only daily_activity needs to be used

```python
sleep = pd.read_csv('/'.join([os.getcwd(),files[-2]]))
```

```python
sleep
```

```python
sleep.rename(columns={'SleepDay': 'ActivityDay'}, inplace=True)
sleep['ActivityDay'] = pd.to_datetime(sleep['ActivityDay'])
sleep['ActivityDay'] = sleep['ActivityDay'].dt.date

daily_activity['ActivityDay'] = pd.to_datetime(daily_activity['ActivityDay'])
daily_activity['ActivityDay'] = daily_activity['ActivityDay'].dt.date
```

```python
daily = pd.merge(daily_activity, sleep, how='left', on=['Id', 'ActivityDay'])
daily
```

```python
daily_activity_grouped = daily_activity.groupby(by=['Id']).mean().reset_index()
```

```python
sleep_grouped = sleep.groupby(by=['Id']).mean().reset_index()
```

```python
daily_grouped = pd.merge(daily_activity_grouped, sleep_grouped, how='left', on=['Id'])
daily_grouped
```
